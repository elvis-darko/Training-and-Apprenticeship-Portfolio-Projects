```python
# Load train and check sample

data = pd.read_csv("C:/Users/elvis_d/DATA_ANALYTICS/GITHUB/Training-and-Apprenticeship-Portfolio-Projects/SENTIMENT ANALYSIS USING HUGGINGFACE/data/Train.csv")

data.sample(5, random_state=2)
```

```python
# Check for null values in the data
data.isnull().sum()
```

```python
# Remove null values from the data
data = data[~data.isna().any(axis=1)]
```

```python
# Check for duplicated values in the data
data.duplicated().any()
```

```python
# Split the train data into effective train and evalutaion sets
train, eval = train_test_split(data, test_size=0.2, random_state=9, stratify=data['label'])
```

```python
# Check sample of train data
train.sample(5, random_state=1)
```

```python
# Check sample of evaluation data
train.sample(5, random_state=2)
```

```python
# Print the shape of the newly created data
print(f"new dataframe shapes: train is {train.shape}, eval is {eval.shape}")
```

```python
# Save the new data created to different csv files
train.to_csv("train.csv", index=False)
eval.to_csv("eval.csv", index=False)
```
